## Part A — Multiple Sequence Alignment with Clustal Omega

We'll align the **insulin gene** from 4 different species and see how similar (or different) it is across the animal kingdom.


### Step 1 — Open Clustal Omega

Go to:

👉 **https://www.ebi.ac.uk/jdispatcher/msa/clustalo**

You'll see a big text box at the top labeled **"Enter or paste a set of sequences..."**


### Step 2 — Paste these 4 insulin sequences

Copy the entire block below (all 4 records together) and paste it into the Clustal Omega box:

```
>Human_INS
ATGGCCCTGTGGATGCGCCTCCTGCCCCTGCTGGCGCTGCTGGCCCTCTGGGGACCTGAC
CCAGCCGCAGCCTTTGTGAACCAACACCTGTGCGGCTCACACCTGGTGGAAGCTCTCTAC
CTAGTGTGCGGGGAACGAGGCTTCTTCTACACACCCAAGACCCGCCGGGAGGCAGAGGAC
>Mouse_Ins1
ATGGCCCTGTTGGTGCTGTTCCTGCCCCTGCTGGCCCTGCTTGCCCTCTGGGAGCCCAAG
CCTGCCCAGGCTTTTGTCAAACAGCACCTTTGTGGTTCCCACCTGGTGGAGGCTCTCTAC
CTGGTGTGTGGGGAGCGTGGCTTCTTCTACACACCCAAGTCCCGCCGTGAAGTGGAGGAC
>Cow_INS
ATGGCCCTGTGGACACGCCTCCTGCCCCTGCTGGCCCTGCTGGCGCTCTGGCCCCCAGCC
CCTGCCCAGGCCTTCGTCAACCAGCACCTGTGCGGCTCCCACCTGGTGGAAGCTCTCTAC
CTGGTGTGCGGGGAACGCGGCTTCTTCTACACGCCCAAGGCCCGCCGGGAGGTGGAGGGT
>Chicken_INS
ATGGCTCTCTGGATCCGATCACTGCCTCTTCTGGCCCTGCTTGTCTTCTCAGGGCCAGGA
ACCAGTTACGCAGCAGCCAACCAACATCTGTGTGGCTCTCACCTGGTGGAAGCTTTGTAC
CTGGTGTGCGGAGAGCGAGGCTTCTTCTACAGCCCCAAGGCCCGCCGCGACGTGGAGCAG
```


### Step 3 — Settings

Below the text box, you'll see settings:

- **Set sequence type** to **DNA**
- Leave the other defaults as they are

Then click the **Submit** button at the bottom.


### Step 4 — Wait for the result

It usually takes **15–30 seconds**. The page will refresh and show your alignment.

*This is the same alignment that has been done thousands of times in research papers — you just ran it yourself.*


### Step 5 — Read the alignment

On the result page, look for the **Alignments** tab. You'll see something like:

```
Human_INS    ATGGCCCTGTGGATGCGCCTCCTGCCCCTGCTGGCGCTGCTGGCCCTCTGGGGACCTGAC
Mouse_Ins1   ATGGCCCTGTTGGTGCTGTTCCTGCCCCTGCTGGCCCTGCTTGCCCTCTGGGAGCCCAAG
Cow_INS      ATGGCCCTGTGGACACGCCTCCTGCCCCTGCTGGCCCTGCTGGCGCTCTGGCCCCCAGCC
Chicken_INS  ATGGCTCTCTGGATCCGATCACTGCCTCTTCTGGCCCTGCTTGTCTTCTCAGGGCCAGGA
             *****  ** *  *    *  ** ** ******* ** ** ** ** *      
```

Below each column, you'll see one of three symbols:

- `*` = **identical** in all 4 species (conserved)
- `:` = **strongly similar** (similar amino acid type)
- `.` = **somewhat similar**
- ` ` (blank) = **different**


### Step 6 — Reflection questions

Look carefully at your alignment and answer:

1. **Which species is most similar to humans** in this region — mouse, cow, or chicken?
2. **Which is most different?**
3. **What % of positions appear identical (with `*`) across all 4 species?**
4. **Why do you think the START of the gene has more matches than the middle?**

*Hint for #4: the start (ATG) is the start codon — it never changes across species.*


### Step 7 — Click "Result Summary"

On the same results page, find a tab labeled **Result Summary** or **Phylogenetic Tree**.

**Look at the tree** — it shows which species are most closely related, based on this gene alone.

Insulin tells a real evolutionary story: the species that are closer on the tree are closer biological relatives.

## Part B — Look Up a Real Disease Variant in ClinVar

Now we shift from "how species differ" to "how individuals differ" — and how some of those differences cause disease.

We'll look at a real BRCA1 variant that causes hereditary breast cancer.


### Step 1 — Open ClinVar

Go to:

👉 **https://www.ncbi.nlm.nih.gov/clinvar**

ClinVar is NCBI's database of variants linked to human disease — over 3 million entries.


### Step 2 — Search for BRCA1

Type **BRCA1** in the search box and press Enter.

You'll see thousands of results — every recorded BRCA1 variant in the world.

**On the left side**, you'll see filters. Click the filter for **Pathogenic** under "Clinical significance."


### Step 3 — Click any pathogenic variant

**"Pathogenic"** = scientifically confirmed to cause disease.

Click any of these entries — it will open the variant's full record.


### Step 4 — Read the variant page

On the variant detail page, look for these sections:

1. **Variation** — the exact change (e.g., `c.5266dup` means a duplication of one base at position 5266)
2. **Conditions** — what disease this variant causes (e.g., "Hereditary breast and ovarian cancer")
3. **Clinical significance** — how confident scientists are this variant is pathogenic
4. **Genomic position** — chromosome 17, exact base position

**These are the same fields you'd see in a VCF file.**


### Step 5 — Note the details

For your chosen variant, write down:

- **Variant ID** (e.g., VCV000055636)
- **Gene** (BRCA1)
- **Position on chromosome 17** (e.g., 43093454)
- **REF and ALT bases** (the change)
- **Disease it causes**

We'll use these in Part 2 when we read variants from a VCF file.
