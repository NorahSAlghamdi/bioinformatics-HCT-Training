## Exercise 7 — Your first function: GC content
**GC content** is the percentage of G and C bases in a sequence.
Let's write a reusable function to calculate it.


```python
def gc_content(sequence):
    """Return the GC content of a DNA sequence as a percentage."""
    g = sequence.count("G")
    c = sequence.count("C")
    return (g + c) / len(sequence) * 100

# Test it
print("GC% for ATGC:", gc_content("ATGC"))
print("GC% for AAAAATTTTT:", gc_content("AAAAATTTTT"))
print("GC% for GCGCGCGCGC:", gc_content("GCGCGCGCGC"))
```

## Exercise 12 — Find ALL positions of a motif
This is a common pattern in bioinformatics: find every place a motif occurs.


```python
dna = "ATGCGATGCCCATG"
motif = "ATG"

positions = []
start = 0
while True:
    pos = dna.find(motif, start)
    if pos == -1:
        break
    positions.append(pos)
    start = pos + 1

print("ATG is at positions:", positions)
```

## Exercise 14 — Count differences between sequences
When two sequences are the same length, we can count how many positions differ.


```python
seq1 = "ATGGATTTATCTGCTCTT"
seq2 = "ATGGATTTATCTGCACTT"

differences = 0
for i in range(len(seq1)):
    if seq1[i] != seq2[i]:
        differences += 1

print("Differences:", differences)
```

This number is called the **Hamming distance** — a basic measure of sequence similarity.

### ✏️ Your Turn
Write a function that calculates Hamming distance and test it on any two sequences.


```python
def hamming(a, b):
    """Count the number of positions where two sequences differ."""
    diff = 0
    for i in range(len(a)):
        if a[i] != b[i]:
            diff += 1
    return diff

# Test it
print(hamming("ATGC", "ATCC"))  # Should print 1
```

## Exercise 15 — AT content
If GC content is the % of G and C, then AT content is the % of A and T.
These two always add up to 100%.


```python
# Write an AT content function
def at_content(sequence):
    """Return the AT content of a DNA sequence as a percentage."""
    sequence = sequence.upper()
    a = sequence.count("A")
    t = sequence.count("T")
    return (a + t) / len(sequence) * 100

dna = "ATGGATTTATCTGCTCTTCGCGTTGAAGAAGTACAA"
print("AT content:", at_content(dna), "%")
print("GC content:", gc_content(dna), "%")
print("Total:      ", at_content(dna) + gc_content(dna), "%")
```

## Exercise 16 — Build a summary report
Let's put everything together. Write a single function that returns a complete summary of any DNA sequence.


```python
def analyze(sequence):
    """Print a full analysis of a DNA sequence."""
    sequence = sequence.upper()
    print("Length:     ", len(sequence), "bp")
    print("A:          ", sequence.count("A"))
    print("T:          ", sequence.count("T"))
    print("G:          ", sequence.count("G"))
    print("C:          ", sequence.count("C"))
    print("GC content: ", round(gc_content(sequence), 2), "%")
    print("AT content: ", round(at_content(sequence), 2), "%")
    print("ATG motifs: ", sequence.count("ATG"))
    print("TAA motifs: ", sequence.count("TAA"))

# Try it
brca1 = "ATGGATTTATCTGCTCTTCGCGTTGAAGAAGTACAAAATGTCATTAATGCTATGCAGAAAATCTTAGAGTGTCCCATCTGTCTGGAGTTGATCAAGGAACCTGTCTCCACAAAGTGTGACCACATATTTTGCAAATTTTGC"
analyze(brca1)
```
