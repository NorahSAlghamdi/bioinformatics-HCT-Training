## Part A — Reading a FastQC Report

**FastQC** is the most common quality control tool in bioinformatics. After every sequencing run, the first thing a bioinformatician does is open the FastQC report.


### Step 1 — Open the example reports

Go to:

👉 **https://www.bioinformatics.babraham.ac.uk/projects/fastqc/**

On that page, scroll down and click **Examples** in the left sidebar. You'll find two reports linked there:

- **Good Illumina Data**
- **Bad Illumina Data**

Open both — each in its own browser tab. We'll compare them.


### Step 2 — Look at the summary panel (left side)

On both reports, the **left panel** shows a checklist. Each check is marked green ✓, yellow ⚠, or red ✗.

**For the GOOD report:** how many green checks? How many red?

**For the BAD report:** what's different? What's red that wasn't red before?

*Take notes — write down 2 things that turned red in the bad report.*


### Step 3 — Click on "Per base sequence quality"

This is the most important plot. It shows quality across positions in the read.

**On the good report:**
- The yellow boxes stay HIGH (in the green zone)
- The red line stays HIGH
- Quality is consistent across the whole read

**On the bad report:**
- The boxes start dropping into the yellow and red zones
- Especially toward the END of the read
- This is a classic pattern — **read quality drops near the end**


### Step 4 — Click on "Adapter Content"

Adapters are short sequences the machine adds during sequencing. They have to be removed before analysis.

**On the good report:** flat — no adapter contamination.

**On the bad report:** rising curves at the end — adapters showing through. This data needs to be **trimmed** before use.


### Step 5 — Click on "Per sequence GC content"

This shows the distribution of GC content across all reads. A healthy sample shows a smooth, single peak.

If you see **two peaks**, you might have **contamination** — DNA from another organism mixed in.


### Reflection

Take a moment and think:

1. **If you got the BAD report from your lab, what would you do?** (Hint: trim the reads.)
2. **Why is QC the FIRST step before any other analysis?**
3. **What do you think happens if you skip QC and run alignment on bad data?**

*The whole point of QC: catching problems EARLY, before they ruin everything that comes after.*

## Part B — Identifying a Mystery Sequence with BLAST

**BLAST** (Basic Local Alignment Search Tool) is the most-used tool in bioinformatics. It takes any sequence and searches the entire NCBI database for matches.

We'll BLAST a mystery DNA sequence — and see if BLAST can tell us what it is.


### Step 1 — Open NCBI BLAST

Go to:

👉 **https://blast.ncbi.nlm.nih.gov**

You'll see four big colored boxes — one for each type of BLAST. Click **Nucleotide BLAST (blastn)**.

*We're using blastn because we have DNA, and we want to search DNA.*


### Step 2 — Paste this mystery sequence

Copy the sequence below and paste it into the big text box at the top of the BLAST page ("Enter Query Sequence"):

```
ATGGCCCTGTGGATGCGCCTCCTGCCCCTGCTGGCGCTGCTGGCCCTCTGGGGACCTGAC
CCAGCCGCAGCCTTTGTGAACCAACACCTGTGCGGCTCACACCTGGTGGAAGCTCTCTAC
CTAGTGTGCGGGGAACGAGGCTTCTTCTACACACCCAAGACCCGCCGGGAGGCAGAGGAC
CTGCAGGTGGGGCAGGTGGAGCTGGGCGGGGGCCCTGGTGCAGGCAGCCTGCAGCCC
```

*(Don't try to guess what it is — let BLAST tell you.)*


### Step 3 — Choose the database

Below the text box, find the **Database** dropdown. Make sure it's set to:

**Nucleotide collection (nr/nt)**

This is the largest public DNA database — your sequence will be compared against billions of others.


### Step 4 — Click the big BLAST button

Scroll to the bottom and click **BLAST**.

It'll take **30 seconds to 1 minute**. While you wait, read this:

> *"BLAST is doing what no human could ever do — comparing your sequence to billions of others, scoring each match, and ranking them. This used to take days before BLAST was invented in 1990."*


### Step 5 — Read the results

When BLAST finishes, you'll see a results page with several sections.

Scroll down to the **"Descriptions"** section — it shows a TABLE of hits, sorted by score.

**Look at the top hit:**

1. What's the description? (probably something like "insulin")
2. What organism is it from?
3. What's the **E value**? (smaller = more significant — < 0.001 is great)
4. What's the **Per. Ident** %? (higher = closer match — 100% means identical)

**Congratulations — you just identified an unknown sequence using the same tool used in millions of research papers.**


### Step 6 — Click into the top hit

Click on the **Description** link of the top hit. It takes you to NCBI's full record for that gene — the same kind of page you saw last weekend.

**You went from a string of letters → to a real, named, well-studied gene** — in under 2 minutes.
