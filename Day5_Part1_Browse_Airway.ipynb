## 3. Subsetting to the classic 8-sample comparison

The original Himes 2014 paper used a **2 × 2 design**: each cell line was treated with 4 things —

| Treatment | What it is | Mechanism |
|---|---|---|
| Untreated | Control (vehicle only) | — |
| Dexamethasone | Synthetic glucocorticoid (anti-inflammatory) | Binds NR3C1 → nuclear translocation |
| Albuterol | β2-adrenergic agonist (bronchodilator) | Relaxes airway smooth muscle |
| Albuterol + Dex | Combination | Two mechanisms at once |

For our class — and for the **classic Bioconductor `airway` teaching dataset** — we keep only the **Untreated vs Dexamethasone** rows. This is the cleanest two-group comparison and avoids confounding with the bronchodilator effect.

From the 4 donors who have an untreated/dex pair, we use **3 donors** (6 samples total) — keeping things light enough to run instantly in Colab while still being a real 3 vs 3 differential expression comparison.

The 6 SRR runs we will use:

| Run | Cell line | Treatment |
|---|---|---|
| SRR1039512 | N052611 | Untreated |
| SRR1039513 | N052611 | Dexamethasone |
| SRR1039516 | N080611 | Untreated |
| SRR1039517 | N080611 | Dexamethasone |
| SRR1039520 | N061011 | Untreated |
| SRR1039521 | N061011 | Dexamethasone |

## 4. Reproducible reanalysis: Recount2 / Recount3

Raw FASTQ files for this study total **~50 GB** (the SRA page even tells you: ~32 GB for the runs, plus ZQ/SRA copies). Far too big to re-align in Colab. Instead, the community provides reprocessed count matrices:

| Resource | URL | What it gives you |
|---|---|---|
| **Recount2** | https://jhubiostatistics.shinyapps.io/recount/ | Counts re-aligned to hg38 with a uniform pipeline |
| **Recount3** | https://rna.recount.bio/ | Same idea, expanded (>700,000 samples) |
| **ARCHS4** | https://maayanlab.cloud/archs4/ | Counts re-aligned to hg38 with kallisto |

Search Recount2 for `SRP033351`. You should find all 16 airway samples processed.
